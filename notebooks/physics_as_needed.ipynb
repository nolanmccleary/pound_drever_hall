## 1.1: Playing With Wiener


The laser's instantaneous phase $\phi(t)$ tells us how much its electric field has rotated compared to a perfectly stable reference. If the laser were perfectly monochromatic $\phi(t)$ value would increase linearly with time:

$$
\phi(t) = 2\pi f_0t
$$

Where $f_0$ is the carrier frequency.

Realistically, the laser experiences jitter due to noise in the laser cavity, current or temperature. We introduce a term $\chi(t)$ to encode all random fluctuations:


$$
\phi(t) = 2\pi f_0 t + \chi(t)
$$

Where $\chi(t)$ is a Wiener Process, thus it satisfies the condition:
$$
d\chi(t) = \sqrt{2D_\phi}dW(t)
$$

Where $W(t)$ is the Wiener and $D_\phi(t)$ is the rate that it integrates white noise as a function of phase (phase diffusion rate).

This means for any lag $\tau$:

$\Delta\chi(t, \tau) \equiv \chi(t + \tau) - \chi(t) ~ N(0, 2D_\phi|\tau|)$

We can represent our laser via the complex analytic field:
$$
E(t) = Ae^{j\phi(t)}
$$

After autocorrelation:
$$
R_E(\tau) = \langle E^*(t)E(t + \tau) \rangle = A^2e^{j2\pi f_0\tau} \langle e^{j[\chi(t + \tau) - \chi(t)]} \rangle
$$

Since $\Delta\chi$ is zero-mean Gaussian with variance $2D_\phi|\tau|$, we can use the Gaussian characteristic function. Thus:
$$
R_E(\tau) = A^2e^{j2\pi f_0\tau}e^{-D_\phi|\tau|}
$$

By the Wiener-Kinchin Theorem, the power spectral density (PSD) is the Fourier transform of $R_e(\tau)$
$$
S_E(f) = \int_{-\infty}^\infty R_E(\tau)e^{-j2\pi f\tau}d\tau = \int_{-\infty}^\infty e^{-D_\phi|\tau|}e^{-j2\pi(f - f_0)\tau}d\tau = A^2\frac{2D_\phi}{D_\phi^2 + (2\pi(f - f_0))^2}
$$

Given that $e^{-a|\tau|}$ is the standard Lorentzian kernel and substituting $a = D_\phi$.

This is all to say that if you graphed the laser's power dissipation as a function of frequency, you would get a spike (Lorentzian line). And the less noise you have the spikier that spike would be. 

The more intuitive way to think about it is that if we time-time shift a copy of our signal forward, the Wiener process will add some noise to it. If we go far enough forward in time there will be so much noise that the two signals would be completely uncorrelated because $\Delta\chi(t, \tau) \equiv \chi(t + \tau) - \chi(t) ~ N(0, 2D_\phi|\tau|)$ has a variance that scales with $\tau$. A high variance with respect to the difference between the two signals indicates that the covariance between them would be extremely minimal. Since both signals are sinusoids, their correlation integral is also sinusoidal with respect to phase shift and thus tau. This is because two different sinusoids with some given phase shift will correlate more or less depending on that phase shift, and since $\phi(t)$ is a function of time, $\tau$ will induce a phase shift. However, because larger $\tau$ values lead to less overall correlation, that means that our correlation function takes the form of a decaying sinusoid with respect to tau i.e. 

$$
R_E(\tau) = A^2e^{j2\pi f_0\tau}e^{-D_\phi|\tau|}
$$

Since it's a decaying sinusoid, we know two things:

1. It will converge
2. It isn't necessarily a pure sinusoid i.e. there could be other frequency components rather than just $f_0$

Taking the Fourier transform would let us figure out what frequency components contribute most to our laser signal. Frequencies that correlate best with our correlation waveform would also by definition of the semantics behind how correlation is defined relative to power contribute the most to the overall power of our signal. 

$$
S_E(f) = \int_{-\infty}^\infty R_E(\tau)e^{-j2\pi f\tau}d\tau = \int_{-\infty}^\infty e^{-D_\phi|\tau|}e^{-j2\pi(f - f_0)\tau}d\tau = A^2\frac{2D_\phi}{D_\phi^2 + (2\pi(f - f_0))^2}
$$

The overall power stays constant but with more noise becomes spread out among different frequency components. The whole point of our feedback loop is to make this Lorentzian spike more spiky.


## 1.2 Laser Physics

In the section above we made a super basic model for a laser's phase like so:

$$
\phi(t) = 2\pi f_0 t + \chi(t)
$$

It's easy to see why frequency as a function of time is equal to the derivative of phase whith respect to time. If phase is radians as a function of seconds, then it's derivative would be radians per second as a function of seconds:

$$
\omega(t) = \frac{d\phi(t)}{dt} = 2\pi f_0 + \frac{d\chi(t)}{dt}
$$

While the derivative $\frac{d\chi(t)}{dt}$ is not necessarily analytic, we can approximate it numerically. As expected, this means our frequency function is a random process as well.

The very dumbed-down physics explanation for why this happens is as follows:

1. Spontaneous emission causes atoms to drop down to a lower energy level by randomly emitting a photon. This photon has a random phase. Laser fields tend to have all their photons in-phase, that's kind of the point. So this random photon will perturb the phase of the overall field. This gives us our $\chi$. 
2. As per the above, this $\chi$ also gives us a temporary $\Delta\chi$ (not the frat) which temporarily perturbs the frequency of our field. 
3. Given enough random photons, we experience persistent frequency drift. 

This gives us an interesting insight:
Frequency stabilization is fundamentally phase noise suppression. 

It also leads us to an interesting realization:
PDH is cool because it lets us bully quantum physics into submission through the power of recursion.
